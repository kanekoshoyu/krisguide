# Introduction
There are several methods of conducting signboard recognition. One very "buzzword" solution is to directly use Convolutional Neural Network to extract all the features and directly store into the database for matching. The problem is that direct use of CNN takes hundreds, if not thousands of image samples to train an accurate model. We cannot go to Changi Airport just to take all these signage pictures. Also, the computation is too heavy, topping up your AWS bill too expensive to sustain.

As for logo extraction, SIFT is a great pattern extraction and mapping algorithm that are used extensively. However, we cannot use SIFT for this case because SIFT is patent protected, and kicked out of the OpenCV libraries. There are several other logo extraction algorithms, like KAZE and ORB. Unfortunately, after countless testing, it still failed to perform a highly accurate icon recognition. KAZE and ORB requires a high complexity of the features to function, as there are too many other matching features in the sample pictures. It was concluded that CNN is necessary, but the training part has to be minimised with some tricks.

When building a model for computer vision, we should always remember what are the assumptions we can leverage to minimise the computation. Airport signboards are very standardised if you look closely, and we should get the pattens out. From the sample images gathered below, we can summarise some patterns.

1. All Signage has plain surface.
2. Background of signage is black, texts are lit.
3. Signage has a constant horizontal rectangle shape.
4. Each row of signage has the format of DIRECTION, ICON, TEXT(ENGLISH), TRANSLATIONS(OPTIONAL).
5. DIRECTION, ICON, TEXT(ENGLISH) are either blue or yellow

From these 5 patterns of the signage, we can build a signage recognition algorithm.

# Step 0: Upload Local Image to AWS S3 (Storage System)
Be it the final app is in the form of a camera or AR app, image has to be stored in the AWS S3 in order to be processed in AWS Rekognition, the machine-learning based image recognition service.

```python
from PIL import Image, ImageFilter, ImageEnhance
from IPython.display import Image as displayImage
from IPython.display import display
import cv2
import numpy as np
from matplotlib import pyplot as plt
import imutils
import boto3

selected = 'sign2'
url = 'img/raw/'+selected+'.png'

s3 = boto3.resource('s3')
data = open(url, 'rb')
print("Uploading Image, takes up to 10 secs")
img = cv2.imread(url)
s3.Bucket('krisguideimages').put_object(Key='test.png', Body=data)
display(displayImage(url))
print("Uploaded Image")
```

# Step 1: AWS Signboard "Rekognition"
AWS Rekognition is a prebuilt image classifier/recogniser that can identify million types of objects in the images thanks to the prebuilt deep neural network. It is used here to recognise the position of the signboard, minimising the scope of computation on the images.

```python
import json

client = boto3.client('rekognition')

response = client.detect_labels(
    Image={
        'S3Object': {
            'Bucket': 'krisguideimages',
            'Name': 'test.png',
        },
    },
)

NOSCOREBOARD=1
for item in response['Labels']:
    if item['Name']=="Scoreboard":
        NOSCOREBOARD=0
        w = item['Instances'][0]['BoundingBox']['Width']
        h = item['Instances'][0]['BoundingBox']['Height']
        l = item['Instances'][0]['BoundingBox']['Left']
        t = item['Instances'][0]['BoundingBox']['Top']
        c = item['Instances'][0]['Confidence']
        s = "%5.2f" % c
        print("Signboard detected, confidence level: "+s+"%")
        
height, width, channels = img.shape

print(width, height)
print(w,h,l,t)
x1 = int(width*l)
y1 = int(height*t)
x2 = int(width*(l+w))
y2 = int(height*(t+h))

crp = img[y1:y2, x1:x2]

cv2.imwrite('img/processed/crp.png',crp)
display(displayImage('img/processed/crp.png'))


if NOSCOREBOARD:
    print("No Signboard detected")
```

# Step 2: Image Preprocessing
Since pictures are often over/underexposed, histogram equalisation is adopted to normalise the intensity of the images. Several Operations are made to extract the features of the images.

- Histogram Equalisation
- Thresholding (Extract signboard shape&orientation)
- Edge Dilation and Erosion

<img src ='img/res/color_histogram_plot.png' style="width: 50%; height: 50%">

However, It was incapable of recognising the black signage background for some image samples, because the background was also dark, and could not obtain the edge/contour. To cater this issue, two thresholds were used to obtain the features. This helps recognise both under/over-exposed corners of the signbard to be recognised.

- Image Preprocessing
- Warp Signage to front facing image
- Conduct OCR, recognise texts and its bounding box position/orientation

From testing, it was found that adaptive threshold was not a good way of thresholding the features of the signboard, because the signboard contained bright texts which affected the local area averate brightness processed by the kernel.

Edge detection was also tested and found to be useless in this case, as the histogram equalisation has already emphasised the features of the signboard, making it capable of extracting the features directly.

```python
img = cv2.imread('img/processed/crp.png',0)#image selected
imgCol = cv2.imread('img/processed/crp.png')#image selected
imgCpy = img

hist,bins = np.histogram(img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.subplot(1,2,1)
plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.title("Original")

cdf_m = np.ma.masked_equal(cdf,0)
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
cdf = np.ma.filled(cdf_m,0).astype('uint8')
equ = cv2.equalizeHist(imgCpy)
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.subplot(1,2,2)
plt.plot(cdf_normalized, color = 'b')
plt.hist(equ.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.title("Normalised")
plt.show()
cv2.imwrite('img/processed/equ.png',equ)

##############################
equ = cv2.medianBlur(equ,3)
cv2.imwrite('img/processed/equ.png',equ)
display(displayImage('img/processed/equ.png'))


thr_val = 135
ret, thr = cv2.threshold(equ, thr_val, 255, cv2.THRESH_BINARY)
#thr = cv2.adaptiveThreshold(equ,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,5)

neg = 255-thr

kernel = np.ones((2,2), np.uint8) 
neg = cv2.erode(neg,kernel,iterations = 1)

kernel = np.ones((5,5), np.uint8) 
dil = cv2.dilate(neg, kernel, iterations=1) 

kernel = np.ones((3,3), np.uint8) 
ero = cv2.erode(dil,kernel,iterations = 1)
cv2.imwrite('img/processed/ero.png', ero)



#display(displayImage('img/processed/equ.png', height=50, unconfined=True))
display(displayImage('img/processed/ero.png'))
```

2 approaches to detecting rectangles:
- Contour Detection: Find the contour 
- Hough Line Detection: find straight lines, detect 4 intersecting points leading to a rectangle

Contour Detection as shown above is made possible in our use case thanks to the Step 1 and 2. This is because area of the background of the signboard is minimised and the Histogram Equalisation is carried out, highlighting the features of the signboard. 

Hough Line is another common method of finding the rectangle corners. It probablistically find the best fit lines in the whole picture, then finds 4 junctions for a rectangle. problem with this solution is that quite often the silhoutette of the signboard is not in the shape of perfect rectangle, with some diconnected contours in the singboard. Although it can set some values to cover these flaws, inappropriate setting of the parameters often leads to moo many or less hough lines, leading to likelihood in misalignment.

```python
def transform_by4(img, points):
	""" 4 """
	points = sorted(points, key=lambda x:x[1])  # y
	top = sorted(points[:2], key=lambda x:x[0])  # x
	bottom = sorted(points[2:], key=lambda x:x[0], reverse=True)  # x
	points = np.array(top + bottom, dtype='float32')  # 

	width = max(np.sqrt(((points[0][0]-points[2][0])**2)*2), np.sqrt(((points[1][0]-points[3][0])**2)*2))
	height = max(np.sqrt(((points[0][1]-points[2][1])**2)*2), np.sqrt(((points[1][1]-points[3][1])**2)*2))

	dst = np.array([
			np.array([0, 0]),
			np.array([width-1, 0]),
			np.array([width-1, height-1]),
			np.array([0, height-1]),
			], np.float32)

	trans = cv2.getPerspectiveTransform(points, dst)  # 
	return cv2.warpPerspective(img, trans, (int(width), int(height)))  # 
'''
dil2 = 255-dil

cv2.imwrite('img/processed/dil2.png', dil2)
display(displayImage('img/processed/dil2.png'))
'''


cnts = cv2.findContours(ero, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0]  # 
cnts.sort(key=cv2.contourArea, reverse=True)  # 

flag=0
lines=imgCol.copy()
for i, c in enumerate(cnts):
    arclen = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * arclen, True)

    level = 1 - float(i) / len(cnts)  # 
    if len(approx) == 4:
        cv2.drawContours(lines, [approx], -1, (0, 0, 255 * level), 2)
        if flag == 0:#No array formed
            warp = approx.copy()  # warp
            flag=1
        else:
            cv2.drawContours(lines, [approx], -1, (0, 255 * level, 0), 2)
    for pos in approx:
        cv2.circle(lines, tuple(pos[0]), 4, (255 * level, 0, 0))

cv2.imwrite('img/processed/lin.png', lines)
display(displayImage('img/processed/lin.png'))
    
if flag != 0:
    war = transform_by4(imgCol, warp[:, 0, :])  # warp
    cv2.imwrite('img/processed/war.png', war)
    display(displayImage('img/processed/war.png'))
else:
    print("No Contour detected")
```

# Step 3: AWS Rekognition Text in Image (OCR)
OCR is readily available in the AWS Rekognition Package.

```python
s3 = boto3.resource('s3')
warpURL = 'img/processed/war.png'

data = open(warpURL, 'rb')
print("Uploading image, takes up to 10 secs, in real case it will be directly accessed by AWS Rekognition without upload")
img = cv2.imread(warpURL)
s3.Bucket('krisguideimages').put_object(Key='warp.png', Body=data)
#display(displayImage(warpURL))

string = "";
print("Uploaded image, recognised text will be shown below in 2 secs:")
client = boto3.client('rekognition')
response = client.detect_text(Image={'S3Object':{'Bucket':'krisguideimages', 'Name':'warp.png'}})
#print(response)
for item in response['TextDetections']:
    if item['Type']=='LINE' and item['Confidence']>90.0:
        string = string+'\n'+item['DetectedText']

print(string)
```

# Step 4: Icon Recognition with Google AutoML Vision (Building own CNN)
While AWS Rekognition has a nice image recognition package, its still cannot recognise icons and arrows in the signboard. For that, we need to build our own Convolutional Neural Network (CNN). Usually CNN at accuracy of 95%+ is built with image set size in thousandth order. however, as we have already warped the signage to be front facing, we can limit the size of the icon recognition at order of tenths. Building CNN with Vision AI is very straight forward.

First, icons and arrows of the signboard is collected from Google Image Search screenshots. these icons are categorised by its name.

<img src ='img/res/dataset.png' style="width: 50%; height: 50%">

Then, the images are loaded to the Google AutoML Vision (Object Detection) and text labels are assigned. These datasets will automatically be divided into 8:1:1 ratio(Train,Test,Validate).

<img src ='img/res/automl.png' style="width: 50%; height: 50%">

After that, press the "train model button" and wait for about an hour for training the model. 


In thid demo, only the front arrow and lift is modelled because there were insufficient images (less than 10) of the elevators and other arrows.

The model is now ready for deploy in the cloud. It can be accessed with the demo library as shown below. The confidence level of the icon recognition can be further improved by adding more sample images to train the CNN.

# Step 5: Query Signboard ID with its Context

Now that we have both the directions and the facilities listed in the signboard, we can use these data to do a query on its signboard ID. Now the texts obtained from OCR might have several problems, such as misreading some texts and spaces. For that, instead of directly comparing the string, SequenceMatcher library is used. A scale from 0.0 (no match) to 1.0 (fully matched) is derived and compared.

similar("Apple","Appel")<br>
0.8<br>
similar("Apple","Mango")<br>
0.0<br>

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

#similar (string, text)

texts = ["",
         "Arrival Transfer D Flight Info Transit Mall",
         "Skytrain to T1 & T2 Train to City Cash Machine (ATM) Shops & Cafes",
         "Swimming Pool Transfer D Transit Hotel Skytrain to T2 Gates E & F",
         "Skytrain to T2 Gates E & F Skytrain to A15 - A21 Free Singapore Tour SilverKris Lounge Powder Room"
        ]

highest_ci = 0.0
signboardId = -1
for num, text in enumerate(texts):
    ci = similar (string, text)
    if ci>highest_ci:
        highest_ci = ci;
        signboardId = num;


print("Confidence Level:", highest_ci)
print("Signboard ID:", signboardId)

```
